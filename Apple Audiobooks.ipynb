### Apple Audiobooks
In this code, data was fetched from  https://itunes.apple.com/search?term=data&country=us&entity=audiobook&limit=50

```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
```

```python
x = requests.get("https://itunes.apple.com/search?term=data&country=us&entity=audiobook&limit=50")
content = x.content
loaded = json.loads(content)
loaded
```

```python
tea = pd.DataFrame(loaded['results'])
df = tea[["artistName", "collectionName", "releaseDate", "primaryGenreName", "collectionPrice", "collectionViewUrl"]]
df.columns = ['author','collectionName','date','genre','price','url']
df
```

```python
def function1(url):
    response = requests.get(url)
    res = response.content
    soup = BeautifulSoup(res, "html")
    t = soup.find_all("div", class_= "book-badge__value")[0].get_text().strip()
    return t

def function2(url):
    response = requests.get(url)
    res = response.content
    soup = BeautifulSoup(res, "html")
    t = soup.find_all("div", class_= "book-badge__value")[1].get_text()
    return t


x = df["url"].apply(function1)
x1 = pd.DataFrame(x)
x1.columns = ["narrator"]

y = df["url"].apply(function2)
y1 = pd.DataFrame(y)
y1.columns = ["length"]
```

```python
nfp1 = x1.join(y1)
df2 = df.join(nfp1)
df2
```

From here on, I am just cleaning up the dataframe to only have necessary information, as well as some manipulations.

```python
def function3(input):
    return int(input[:4])
t = df2["date"].apply(function3)
def function4(input):
    return float(input)
z = df2["price"].apply(function4)
df2["date"]=t
df2["price"]=z
del df2["url"]
df2
```

Finding the audiobooks of the author that has the largest number of audiobooks

```python
name = df2.groupby("author").count().sort_values(by="collectionName", ascending=False).index[0]
l = df2.loc[df2['author'] == name]
l[["author","collectionName", "genre"]]
```
